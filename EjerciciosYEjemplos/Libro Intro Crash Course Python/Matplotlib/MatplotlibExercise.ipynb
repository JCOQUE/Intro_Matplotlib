```python
from pathlib import Path
import csv
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
```

```python
path = Path('sitka_weather_07-2021_simple.csv')
lines = path.read_text(encoding = 'utf-8').splitlines()

reader = csv.reader(lines)
header_row = next(reader)
print(header_row)
```

```python
col_pos_mapping = {'DATE':2,
                   'TMAX':4, 
                   'TMIN':5,
                  }
```

```python
# Extracting data
highs = list()
dates = list()
for row in reader:
    high = int(row[col_pos_mapping['TMAX']])
    date = str(row[col_pos_mapping['DATE']])
    highs.append(high)
    dates.append(date)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(highs, color = 'r')

ax.set_title("Daily High Temperatures, July 2021")
ax.set_xlabel("")
ax.set_ylabel("Temperature (F)")

plt.show()
```

```python
path = Path('sitka_weather_2021_simple.csv')
lines = path.read_text(encoding = 'utf-8').splitlines()

reader = csv.reader(lines)
header_row = next(reader)

highs = list()
dates = list()
lows = list()
for row in reader:
    date = str(row[col_pos_mapping['DATE']])
    high = int(row[col_pos_mapping['TMAX']])
    low = int(row[col_pos_mapping['TMIN']])
    dates.append(date)
    highs.append(high)
    lows.append(low)
```

```python
fig, ax = plt.subplots(figsize = (10,6))
ax.plot(dates, highs, color = 'r', alpha = 0.5)
ax.plot(dates, lows, color = 'b', alpha = 0.5)

ax.set_title("Daily High and Low Temperatures, July 2021")
ax.set_xlabel("")
ax.set_ylabel("Temperature (F)")
plt.xticks(ticks = np.arange(0, len(dates), len(dates)//9), 
           labels = dates[::len(dates)//9], rotation = 45)

ax.fill_between(dates, highs, lows, fc = 'b', alpha = 0.1)


plt.show()
```
