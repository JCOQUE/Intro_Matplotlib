## Subplots (or Axes)
As we define in the previous notebook, an Axe is a plot. And the relationship between a Figure, an Axe and a Graph is given by the next formula:

$$ Figure > Plot > Graph $$

Basically this is saying that a **plot** may contain multiple **graphs**. And a **figure** may contain multiple **plots**.

## <a id = 'create'>Creating subplots</a>

We've seen in the notebook **'Two styles of coding graphs'** that when plotting multiplte plots, i.e. creating subplots, the OO-style (*with fig and ax as objects*) is more appropiate.

Creating subplots this way can be done in a slow and boring way (one by one) or in a fast and dinamic one (with loop).

It depends on how we have our plotting data stored that we will be using either the former or the latter.
- If our plotting data is stored in a list-like or dict-like or some other iterable object that can be accessed by a loop, **and** our plotting is always the same (all plots are pieCharts, or histograms or plots or scatter...) then we'll go for the fast way.
- Otherwise, we will have to plot our data plot by plot.

### <a id = 'slow'>Slow way: plot by plot</a>

If we cannot iterate through our data then we will have to go plot by plot introducing the data. This is the way we've been doing it until now. If we have 2,3 or 4 plots, it is not such a pain on the neck, but if we had more... it'd be. Let's see what I mean

```python
x = np.array([0,1,2,3,4,5])
y1 = np.array([4,6,5,2,2,5])
y2 = np.array([9,2,3,1,8,7])
y3 = np.array([5,8,9,7,0,2])
y4 = np.array([1,2,3,3,2,1])

# Let's plot 4 subplots in a 2x2 grid
fig = plt.figure()
fig,ax = plt.subplots(nrows = 2, ncols = 2)
ax[0,0].plot(x,y1)
ax[0,1].bar(x,y2)
ax[1,0].scatter(x,y3)
ax[1,1].plot(x,y4)
```

### <a id = 'fast'>Fast way: for loop</a>
If we see each plot as a position in a matrix, we could loop through each axe as if it was a position in a matrix.

```python
fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(3.5, 2.5),
                        layout="constrained")

# looping as if it was a 2x2 matrix
for row in range(2):
    for col in range(2):
        axs[row, col].annotate(f'axs[{row}, {col}]', (0.5, 0.5),
                            transform=axs[row, col].transAxes,
                            ha='center', va='center', fontsize=18,
                            color='darkgrey')
fig.suptitle('plt.subplots()')
```

### <a id = 'extra'> Extra way </a>
There is a way that might be a little faster declaring axis this way in case we only want to plot 2 axis without using for loops. This is passing the axis as a `tuple`.

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(6, 4),
                        layout="constrained")


ax1.set_title('First plot')
ax1.plot(x,y1)

ax2.set_title('Second plot')
ax2.plot(x,y2)


fig.suptitle('axis with tuples')
```

### <a id = 'mosaic'> Subplot mosaic </a>
With this feature, we can customize the layout of our subplots. `plt.subplot_mosaic()` works with nested lists (= nested matrix) as parameter, and strings of **IDs** passed. We'll see this better with an example:

```python
# Create layout with a nested matrix
# We set an ID corresponding to each plot and the position it is
# going to occupy in our plot
layout = [
    ["ID1", "ID1", "ID2"],
    ["ID3", "ID4", "ID4"],
    ["ID3", "ID5", "ID5"]
]

fig, axd = plt.subplot_mosaic(layout, figsize=(10,10))

# Generate some random data

data = np.random.random((10,100)) 

# Plot the data on the subplots

axd['ID1'].imshow(data) # creates a 2-D image
axd['ID1'].set_title('Image Plot')

axd['ID2'].hist(data[0], color='green') # creates a histogram off all the data flattened to a 1-D array
axd['ID2'].set_title('Histogram')

axd['ID3'].scatter(data[1, :], data[2, :], color='purple') # scatter plot with second and third columns 
axd['ID3'].set_title('Scatter plot')

axd['ID4'].plot(data[0], color='orange') # generate a lineplot of the first row of data
axd['ID4'].set_title('Line Plot')

axd['ID5'].bar(np.arange(100), data[0], color='red')
axd['ID5'].set_title('Bar Plot')

# Add main title

fig.suptitle('Matplotlib subplot_mosaic Example', fontsize=16)

# Adjusting the layout for better view
plt.tight_layout()
plt.show()
```
