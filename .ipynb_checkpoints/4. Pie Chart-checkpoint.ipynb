## <a id = 'basic'>Basic Pie Chart</a>
Since we've already seen what a (basic) Pie Chart in the *0.basic_intro_to_matplotlib*, I am not going to explain what some attributes like **explode, shadow, start angle** do.

To make the first **pie plot** more realistic, rather than just plotting some random data, let's give it a meaning. Suppose we do have a dictionary with social media as *keys* and the number of hours spent in that app during the week as *values*.

```python
#made up data
data = {'Youtube': 15, 
        'Instagram': 8,
        'Tik Tok': 7,
        'Whatsapp': 10,
        'Twitter': 2}

colors = ['#FF0000', 'LightPink', 'k', 'Chartreuse', 'DodgerBlue']
social_media_apps = list()
hours_social_media = list()

for k, v in data.items():
    social_media_apps.append(k)
    hours_social_media.append(v)
    
plt.pie(hours_social_media, explode = [0.2, 0, 0, 0, 0],
        shadow = True, startangle = 0, colors = colors)
plt.legend(title = 'Social Media Apps', labels = social_media_apps, loc = 'upper center', 
          bbox_to_anchor = (1.15, 0.7), shadow = True)
```

### <a id = 'pct'>Percentages and its Color </a>
You can add percentages to the PieChart to get a better perspective of data that you have.

- This can be done with the attribute `autopct`. <br><br>

    - `autopct = '%1.1f%%'` is pretty standard
<br><br>
- The **color of the percentage** labels can be modified with the attribute `textprops` which is a dictionary containing **color** as a key.

```python
plt.pie(hours_social_media, explode = [0.2, 0, 0, 0, 0],
        shadow = True, colors = colors,  autopct = '%1.1f%%', textprops = {'color':'w'})
plt.legend(ncol = len(social_media_apps), title = 'Social Media Apps', labels = social_media_apps, loc = 'upper center', 
          bbox_to_anchor = (0.5, 1.2), shadow = True)
```

Another more sophisticated way of plotting percentages would be by declaring a function and passing it to the `autopct` attribute. This way, we can control exactly what to place inside each triangle. 

Let's take a look at the next example that uses a function called `func` to place the percentage with the actual value inside each triangle.

```python
# Creating dataset
cars = ['AUDI', 'BMW', 'FORD', 
        'TESLA', 'JAGUAR', 'MERCEDES']

# car's kilometers 
data = [23, 17, 35, 29, 12, 41]
 

# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)
 
# Creating color parameters
colors = ( "orange", "HoneyDew", "brown",
          "grey", "LightSkyBlue", "Khaki")
 

# function that calculates the percentages based on the sum of all the values
def func(pct, allvalues):
    '''
    pct: single value of the data list
    data: list containing all values
    
    return: a string containing the percentage value and the absolute value
    '''
    absolute = int(pct / 100.*np.sum(allvalues)) # calculates the percentage
    return "{:.1f}%\n({:d} km)".format(pct, absolute)
 
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
ax.pie(data, 
        autopct = lambda pct: func(pct, data),
        explode = explode, 
        labels = cars,
        shadow = True,
        colors = colors,
        startangle = 90,
        textprops = dict(color ="navy"))
 
# Adding legend
ax.legend(cars,
          title ="Cars",
          loc ="center left",
          bbox_to_anchor =(1.1, 0, 0.5, 1))
 
ax.set_title("Customizing pie chart")
 
# show plot
plt.show()
```

### <a id = 'wedge'>Wedgeprops</a>
Customize the borders of the PieChart. There are several ways to customize it, but the most important ones are:
- **linewidth (or lw)**: float
- **edgecolor**: string
- **linestyle (or ls)**: '-', '--', '-.', ':', ''

```python
years = [2016, 2017, 2018, 2019, 2020] 
profit = [15, 19, 35, 14, 17] 
  
# Plotting the pie chart 
plt.pie(profit, labels = years, autopct = '%1.1f%%', 
        startangle = 90,  
        wedgeprops = {"edgecolor" : "black", 
                      'linewidth': 2}) 
  

# Display the graph onto the screen 
plt.show()
```

```python
# the slices are ordered and plotted counter-clockwise: 
sales = ['Product A', 'Product B',  
         'Product C', 'Product D'] 

profit = [20, 30, 25, 20] 
explode = (0.1, 0, 0.1, 0)  
colors = ['LightGreen', 'LightSeaGreen', 'MediumPurple', 'LightGrey']  

plt.pie(profit, explode = explode, labels = sales, 
        autopct = '%1.1f%%',shadow = True, 
        startangle = 90, 
        colors = colors,
        wedgeprops = {"edgecolor":"Magenta", 
                      'lw': 2,
                      'ls':'-.'}) 
  

plt.show()
```

### <a id = 'radius'>Radius</a>
With the attribute `radius` you can customize the size of the PieChart.
- By default  `radius = 1`.

```python
# Example from above

years = [2016, 2017, 2018, 2019, 2020] 
profit = [15, 19, 35, 14, 17] 
  
# Plotting the pie chart 
plt.pie(profit, labels = years, autopct = '%1.1f%%', 
        startangle = 90,  
        wedgeprops = {"edgecolor" : "black", 
                      'linewidth': 2},
       radius = 3) #pretty darn big

  
# Display the graph onto the screen 
plt.show()
```

### <a id = 'swap'>Swap label and autopct text positions</a>

`pctdistance` and `labeldistance` are both attributes of the PieChart with **default values** of 0.6 and 1.2 respectively.

If we wanted to swap label and autopct places, we would just have to swap their respective values.

```python
years = [2016, 2017, 2018, 2019, 2020] 
profit = [15, 19, 35, 14, 17] 
  
# Plotting the pie chart 
plt.pie(profit, labels = years, autopct = '%1.1f%%', 
        startangle = 90,  
        wedgeprops = {"edgecolor" : "black", 
                      'linewidth': 2},
        pctdistance = 1.2,
        labeldistance = 0.6) 
  
plt.legend(labels = years, bbox_to_anchor = (1.35, 0.6), loc = 'center right')
# Display the graph onto the screen 
plt.show()
```
