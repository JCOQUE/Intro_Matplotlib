## <a id = 'simple'> Simple Barplots</a>
Let's start for something very simple. A  normal barplot, with 2 bars, each one with a different color.

```python
# Made up data
men_average_height = 1.80
women_average_height = 1.70
labels = ['Men', 'Women']
colors = ['aqua', 'coral']

plt.yticks(np.arange(1.60, 1.90, 0.05))
plt.ylim(1.60, 1.90)
plt.xlabel('Gender')
plt.ylabel('Height')

plt.title('Average Height')

# Plotting bars for men and women separately
plt.bar('Men', men_average_height, color=colors[0], label='Men')
plt.bar('Women', women_average_height, color=colors[1], label='Women')

plt.legend(title='gender')
plt.show()
```

## <a id = 'mult'>Multiple barplots</a>
Imagine we have data of the average heights of men and women from 5 different countries: Spain, France, Portugal, Italy and Germany, and we want to plot them all in one graph.

### Example 1

```python
#made up data
dict_average_heights = {'men': [176, 174, 179, 169, 185],
                        'women': [168, 164, 169, 173, 179]}

gender_labels = ['men', 'women']
country_labels = ['Spain', 'France', 'Portugal', ' Italy', 'Germany']
barWidth = 0.25 #necessary for plotting one graph next to each other
x = np.arange(len(country_labels))

# each loop is going to plot me all heights of each gender. 
# in other words, each loop is going to plot me the bars of a color
for i, gender in enumerate(gender_labels):
    plt.bar(x + (barWidth * i), dict_average_heights[gender], width = barWidth, color = colors[i])

#NOTE:
# x + (barWidth * i) is placing the next color bar (the women's height) next to the men's height.
# that's why we needed the barWidth variable, so gender's heights do not overlap.

plt.xticks(x + (barWidth / 2) ,country_labels) # x + (barWidth / 2) is so that the xtick is in the middle
plt.yticks(np.arange(155, 190, 3))
plt.ylim(155, 190)
plt.legend(title = 'gender', labels = gender_labels)


plt.show()
```

>**Note**: Notice that the for loop was looped *'gender-wise'* and not *'country-wise'*.

### Example 2

```python
barWidth = 0.25
colors = ['r', 'k', 'DodgerBlue']
# made up data
views_per_year = {'Youtube':[43, 37, 48, 56, 76],
                  'Tik Tok': [24, 38, 64, 446, 2034],
                  'Facebook':[27, 24, 19, 21, 13]}
x = np.arange(0, 5, 1)
years_labels = np.arange(2015, 2020, 1)
social_media_apps = list()
for k in views_per_year:
    social_media_apps.append(k)# I add the keys of the dictionary to a list, i.e names of the SM apps
    
for i, social_media in enumerate(social_media_apps):
       plt.bar(x +( barWidth * i), views_per_year[social_media], width = barWidth, color = colors[i])
        

plt.xlabel('Years')
plt.ylabel('Billion views')
plt.title('Views in social media per year')
plt.xticks(x + barWidth, years_labels) # + barWidth is important so that the year label is in the middle. 
                                       # We can see from the last example that denpending on how many bars we have
                                       # we have to add a bit more distance to the x_axis to the right.  
plt.yticks(np.arange(0, 2100, 500))

plt.legend(title = 'Social Media Apps', labels = social_media_apps)
 
plt.show()
```

>**Note**: Notice that the xticks in the graph are accomplished in two steps. First with a simple array [0, 1, 2,...], and then I change it to the one that I actual want. To see it clearer, the steps are:

1. auxiliar xticks (np.arange(0, ... ))
2. I move these ticks with `plt.xticks(x + barWidth)` so that the **xtick is in the middle**.
3. Then I set the xticks that I want.

### Example 2
Let's see now an example with multiple barplots on top of each other.

```python
# create data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])
 
# plot bars in stack manner
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.bar(x, y3, bottom=y1+y2, color='y')
plt.bar(x, y4, bottom=y1+y2+y3, color='g')
plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1", "Rond 2", "Round 3", "Round 4"])
plt.title("Scores by Teams in 4 Rounds")
plt.show()
```

## <a id = 'horiz'>Horizontal Barplots</a>

```python
player_goals = {'Ronaldo': 567,
                'Messi': 601,
                'Pele': 439,
                'Maradona': 401,
                'Zidane': 254,
                'Raul': 385,
                'Mbappe': 202}

#order the dictionary so the barplot is nicer
player_goals = dict(sorted(player_goals.items(), key = lambda x: x[1]))
players_list = list()
goals_list = list()

for k,v in player_goals.items():
    players_list.append(k),
    goals_list.append(v)

x = np.arange(0, 650, 50)
colors = np.random.rand(len(players_list),3) #the 3 needs to be added because of RGB
plt.barh(players_list, goals_list, color = colors)

plt.show()
```

### <a id = 'shoriz'>Stacked horizontal barplots</a>
They work the same way with a tiny change. With vertical plots, we use the attribute `bottom`. 

Here, with horizontal barplots, since they are horizontal, this attributes changes to  `left`.

Let's see the example we did before

```python
# create data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])
 
# plot bars in stack manner
plt.barh(x, y1, color='r')
plt.barh(x, y2, left=y1, color='b') # left instead of bottom
plt.barh(x, y3, left=y1+y2, color='y') # left instead of bottom
plt.barh(x, y4, left=y1+y2+y3, color='g') # left instead of bottom
plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("Scores by Teams in 4 Rounds")
plt.show()
```

## <a id = 'labels'>Adding labels to our barplots</a>
In order to do this, we are going to use `plt.bar_label()` that receives by parameter the `plt.bar()` (or the `ax.bar()` if we are using the OO-style). It works exactly the same with both styles.

Let's reuse some of our plots to see how to add labels.

```python
#made up data
dict_average_heights = {'men': [176, 174, 179, 169, 185],
                        'women': [168, 164, 169, 173, 179]}

gender_labels = ['men', 'women']
country_labels = ['Spain', 'France', 'Portugal', ' Italy', 'Germany']
barWidth = 0.25 #necessary for plotting one graph next to each other
x = np.arange(len(country_labels))


for i, gender in enumerate(gender_labels):
    barplot = plt.bar(x + (barWidth * i), dict_average_heights[gender], width = barWidth, color = colors[i]) # <---------
    plt.bar_label(barplot, labels =dict_average_heights[gender], label_type = 'edge', rotation = 0) # <-----------------

plt.xticks(x + (barWidth / 2) ,country_labels)
plt.yticks(np.arange(155, 190, 3))
plt.ylim(155, 190)
plt.legend(title = 'gender', labels = gender_labels)


plt.show()
```

This `plt.bar_label()` function has the attribute:
- **labels** that receives the data we  want to insert in our barplot
- **label_type**: locate where you want these labels. Possible values (edge, center)
- **padding**: normally used with `label_type = 'edge'`. Sets a distance between the barplot and the label
- **rotation**: Rotate the label the degrees you want.
- **color**: set the color of the label
- **fontsize**: set the fontsize of the label
- **style**: label style. Possible value are: normal (default), italic, oblique

Let's see the same example with other attribute's values:

```python
for i, gender in enumerate(gender_labels):
    # here I pass the attribute height, which is always going to be the same as labels in the next line
    barplot = plt.bar(x + (barWidth * i), height = dict_average_heights[gender], width = barWidth, color = colors[i])
    # you can declare the labels here, in the previous line or in both. 
    plt.bar_label(barplot, labels =dict_average_heights[gender], label_type = 'edge', rotation = 90,
                 color = 'c', fontsize = 15, padding = 20) # <-----------------

plt.xticks(x + (barWidth / 2) ,country_labels)
plt.yticks(np.arange(155, 195, 6))
plt.ylim(155, 195)
plt.legend(ncol = len(gender_labels), labels = gender_labels,
          loc = 'upper left')


plt.show()
```

<br><br>

>**Note**: The ***labels*** attribute passed to `plt.bar_label()` could  be ommitted since it is already passed to `plt.bar()` with the attribute **height**.

### <a id = 'slabels'>Adding labels to a stacked barplot</a>

```python
# create data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])

list_y = [y1, y2, y3, y4]
colors = ['r', 'b', 'y', 'g']
sum_of_y = 0 #variable needed for the stack. I am going to use to sum  y1 + y2... in the bottom attribute

for i, y in enumerate(list_y):
    if i !=  0: #sum y if you  are not the first stack
        sum_of_y += list_y[i-1]
    barplot = plt.bar(x, height = y, bottom = sum_of_y, color = colors[i])
    plt.bar_label(barplot, label_type = 'center', color = 'white', style = 'italic',
                 fontsize = 10, rotation = 0)
    # in this example, I decided not to pass the labels attribute because
    # as I said, it is already been passed in plt.bar with the heigth attribute


# plot bars in stack manner

plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("Scores by Teams in 4 Rounds")
plt.show()
```
