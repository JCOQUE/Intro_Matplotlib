## <a name = 'plot'>Normal plots. Plot()</a>

Now, by default, matplotlib takes a starting point and an endpoint, and draws a straight line between these two points. So matplotlib, by default, doesn't see two points as individuals but as a segment.

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

#draws a line between the points (1,3) to (3,10)
plt.plot(xpoints, ypoints)
plt.show()
```

The same will happen if we had multiple points

```python
x_multiple_points = np.array([8, 11, 71, 82])
y_multiple_points = np.array([3, 8, 1, 10])

plt.plot(x_multiple_points, y_multiple_points)
plt.show()
```

If we don't care about the x-axis values, we can just pass the y-axis values and the $x$ values will take default values (0,1,2,..., n)

```python
plt.plot(y_multiple_points)
plt.show()
```

But, what if we wanted to see the points as individuals? Then we would use the **scatter plot** that we will see later on. But to simplify things now, you can plot individual points with `plt.plot(x,y,'o')`.

```python
plt.plot(x_multiple_points, y_multiple_points, 'o')
plt.show()
```

You can customize the point's shapes

```python
plt.plot(x_multiple_points, y_multiple_points, 'X')
plt.show()
```

You can see all the different shapes in [this link](https://matplotlib.org/stable/api/markers_api.html).

You can add a line and  its color by passing extra arguments to the string.  **shape|line|color**

```python
#o is the shape | : means dotted line | c means cyan color
plt.plot(y_multiple_points, 'o:c')
plt.show()
```

Although this is a very fast and easy way to do it, maybe it is not the most intuitive and easy-readable. Sometimes is better to specify each feature by its name.

### <a name = 'mark'>Marker</a>
You can also make a new mix plot of both plots we have seen with the keyword  **marker**.

```python
plt.plot(x_multiple_points, y_multiple_points, marker='*')
plt.show()
```

You can also add a size to the dots (**ms** for marksize) and an edgecolor (**mec** for mark edge color).

```python
plt.plot(y_multiple_points, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

Finally, you can modify the dot's color with mfc (markerfacecolor)

```python
plt.plot(y_multiple_points, marker = 'o', ms = 20, mec = 'c', mfc = 'y')
plt.show()
```

### <a name = 'ml'> Multiple graphs in one plot</a>
Before entering into this, I'm going to first explain the `plt()` function that we are using all the time.

The way I see **plt()** is like a bag where each time we use it, we are introducing a 'plotting info' into that bag.
And when we finally use the method `plt.show()`, we extract everything that it is the bag **-.plt()-**  and plot that.

So basically, to plot multiple line, we just have to add as much function as we want to the `plt.plot()` method, and finally use `.show()` to plot them all.

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
y2 = np.array([0, 1, 2, 3])
y3 = np.array([6, 2, 7, 11])

plt.plot(x1, y1)# first func
plt.plot(x1,y2)# second func
plt.plot(x1,y3)# third func
plt.show()
```

##### bbox_to_anchor
This concept is really hard to explain, but very useful. It allows you, with `loc` attribute, to place the legend in a more personalized way. [Here](https://jdhao.github.io/2018/01/23/matplotlib-legend-outside-of-axes/#:~:text=In%20plain%20words%2C%20bbox_to_anchor%20accepts,by%20width%20and%20height%20%2C%20respectively.) is a link that explains `bbox_to_anchor` really well. 

One thing that led me to understand what was going on in the link above, was realizing that the red rectangles and red points are static. They are always in the same place. And the `loc` attributes, locates the legend around this rectangle or point (see the graph below).

Another thing we need to understand, is that the coordenates that we pass to `bbox_to_anchor` as parameters, treat the plot as a [0,1]x[0,1] box. That means the coordenates (0,0) would be the **bottom left corner** and the (1,1) coordenates would be the **top right corner**.

- The sentence ***think outside the box*** couldn't fit better in this case; because we can pass coordenates greater than one to the `bbox_to_anchor` attribute in case our legend overlaps our figure (which is common  when plotting PieCharts) or simply because we want our legend to be *outside* from our plot.

>**Note**: `bbox_to_anchor` allows you to pass 4 parameters. If 4 parameters are passed **(x0, y0, height, width), it will create a rectangle. If only two are passed **(x0, y0)**, it will create a point.

Let's see this explanation more visually (don't care about the code; which by the way, is in the above link).

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

locs = ['upper right', 'lower left', 'center left', 'lower center', 'center',
        'right']

x0, y0, width, height = 0.5, 0.5, 0.1, 0.4

x = np.arange(0.1, 4, 0.1)
y = 1.0/x

fig = plt.figure(figsize=(10, 10))

idx = 1
for i in range(0, 2):
    for j in range(0, 3):
        ax = fig.add_subplot(3, 2, idx)
        ax.plot(x, y, label=r'$\frac{1}{x}$')
        ax.legend(loc=locs[idx-1], bbox_to_anchor=(x0, y0, width, height),
            edgecolor='g', fontsize='large', framealpha=0.5,
            borderaxespad=0)
        ax.add_patch(
            patches.Rectangle((x0, y0), width, height, color='r',
                            fill=False, transform=ax.transAxes)
        )
        ax.text(0.6, 0.2, s="loc = '{}'".format(locs[idx-1]),
        transform=ax.transAxes)
        idx += 1

fig.suptitle('4 parameters -> rectangle')
plt.show()
```

## <a name = 'subplots'>Subplots</a>

With the `subplot()` function you can draw multiple plots in one figure.
This function receives 3 parameters:<br>
- **nrows** the subplot is going to have
<br>
- **ncols** the subplot is going to have
<br>
- **index** the subplot is going to have

We are going to understand this much better with some examples

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# this means that the subplot is going to have:
# - 1 row
# - 2 cols
# - This subplot is going to be located in the first position
plt.subplot(1, 2, 1) 
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

 
# this means that the subplot is going to have:
# - 1 row
# - 2 cols
# - This subplot is going to be located in the second position
plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

Now, how would we set the position if we had more than two rows? It will be the same. 
Give an order to the subplots taking into account it starts counting from top-left and by rows.
This is that a 2x2 grid:
- 1 would be the position at (0,0) (top left)
- 2 would be the position at (0,1) (top right)
- 3 would be the position at (1,0) (bottom left)
- 4 would be the position at (1,1) (bottom right)

>**Note**: We are going to add some title to each as well. The same process would be for $x$ and $y$ labels.

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 2, 1)
plt.title('top left')
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

 
plt.subplot(2, 2, 2)
plt.title('top right')
plt.plot(x,y)

#plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([4, 2, 1, 4])

plt.subplot(2, 2, 3) 
plt.title('bottom left')
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([7, 20, 1, 40])

 
plt.subplot(2, 2, 4)
plt.title('bottom right')
plt.plot(x,y)


plt.show()
```

### <a id ='layout'> Layout </a>
Sometimes, when plotting multiples graphs, it might happen that their titles overlap one another, and we want to separate the graph a little bit. We can do this by setting a layout. There are two main types of layout:
- **constrained**
- **tight layout** (most used)


>**Note**: The layout can be either declared in the [figure](#figure) or in `plt()`.

>**Note**: if declaring it in the figure, the figure englobes all the plots so it must be declared just once (not one for each plot) and at the beginning.

##### Without layout

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) 
plt.title('This is some overlapping title to be fixed')

plt.plot(x,y)

#plot 2:
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.title('This is some other overlapping title')

plt.plot(x,y)


plt.show()
```

##### With layout (tight layout)
You see now the overlapping, right? Well we can fix this by setting layout: 

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) 
plt.title('This is some overlapping title to be fixed')

plt.plot(x,y)

#plot 2:
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.title('This is some other overlapping title')

plt.plot(x,y)

# Added tight layout in plt()
plt.tight_layout()


plt.show()
```

```python
# Now we are going to add the layout in the figure 


#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# Declaring the layout at the beginning and just once
plt.figure(layout = 'tight')

plt.subplot(1, 2, 1)
plt.title('Thsss iss the tittle of the first grah')

plt.plot(x,y)

#plot 2:
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.title('Thsss iss the tittle of the second grah')

plt.plot(x,y)


plt.show()
```

##### constrained layout

```python
# Now we are going to add the layout in the figure 


#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# Declaring the layout at the beginning and just once
plt.figure(layout = 'constrained')

plt.subplot(1, 2, 1)
plt.title('This is some overlapping title to be fixed')

plt.plot(x,y)

#plot 2:
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.title('This is some other overlapping title')

plt.plot(x,y)


plt.show()
```

## <a name = 'piecharts'>Pie Charts</a>
To create a pieChart in MatplotLib, use the function `pie()`.

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

>**Note**: The pie chart draws the pieces from smallest to biggest, starting at the x-axis (or 3:00am).

### <a name = 'pclabels'>Labels</a>
You can add labels by passing a list of strings to the attribute **labels** of `pie()`.

```python
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

>**Note**: When plotting a pieChart, it is a good practice to save data and the respective label in a dictionary so it is easier later to plot it.

### <a name = 'startangle'> Start angle </a>

As mentioned above, the pieChart starts plotting at angle 0 (or 3:00am). If you wanted to change this starting angle for whatever reason you could do it with the attribute **startangle** in `pie()`.



```python
plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

### <a name = 'exp'>Explode</a>
Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.

```python
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

### <a name = 'shad'>Shadow</a>
You can add a shadow to the chart to make it look cooler. This is set with the **boolean** parameter `shadow`.

```python
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

### <a name = 'pc_color'>Color</a>

There are three different ways to set color explained [here](#Colorr) before. 

This could be useful in case we wanted to set specific color to our pie chart.

```python
mycolors = ["black", "hotpink", "b", "#4CAF50"] # here you can see the 3 different ways to set colors

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

### <a name = 'pcleg'>Legend</a>
Additionaly to the labels, we have the legend.
>**Note**: You cannot use `legend()` without using labels.

```python
# with labels
plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

```python
# using legend without labels throws an error
plt.pie(y)
plt.legend()
plt.show()
```

#### <a name = 'leyh'>Legend with header</a>
You can add a title to the legend in case you wanted to do so.

```python
plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:", loc = 'lower left')
plt.show()
```
