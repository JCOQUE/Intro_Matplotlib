## <a id = 'ps'>Pyplot-style (implicit)</a>

This is how we created graphs in the intro notebook in this folder. 

We used `plt()` and we said that this worked like a bag where you added things, and it was when you called `plt.show()` that all these things that you added were displayed. Let's look at two examples:

#### <a id = 'pssg'>Plotting a single graph with Pyplot-style</a>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0,1,2,3,4,5])
y = np.array([7,2,7,1,6,3])

# my graph
plt.figure(figsize = (6,6), layout = 'tight')
plt.xlabel('This is the X label')
plt.ylabel('This is the Y label')
plt.title('Graph with Pyplot-style', loc = 'center')
plt.grid(color = 'green', ls= ':')

plt.plot(x, y, label = 'myGraph')
plt.legend()


plt.show()
```

#### <a id = 'psmg'>Plotting multiple graphs with Pyplot-style</a>
>**Note**: Notice that the command is subplot (without an 's' at the end). We'll see later that for the second way to code a graph, the command is going to be subplot**s** (with an 's' at the end). [Subplot documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html), [Subplots documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html)

```python
y2 = np.array([2,6,1,4,6,7])

# width and height (in that order)
plt.figure(figsize = (6,4), layout = 'tight')
# first graph
plt.subplot(1,2,1) #it's going to have 1 row, 2 cols and be at position 1
plt.xlabel('This is the X label of the first graph')
plt.ylabel('This is the Y label of the first graph')
plt.title('This is the title of the first graph', loc = 'center')
plt.grid(color = 'green', ls= ':')
plt.plot(x, y)

# second graph
plt.subplot(1,2,2) #it's going to have 1 row, 2 cols and be at position 2
plt.xlabel('This is the X label of the second graph')
plt.ylabel('This is the Y label of the second graph')
plt.title('This is the title of the second graph', loc = 'center')
plt.grid(color = 'black', ls= ':')

plt.suptitle('This is the title of the subplot')

plt.plot(x, y2)

plt.show()
```

##  <a id = 'oo'>Object- Oriented style (explicit)</a>
Creates **Figures and Axes**, and call methods on them

>**Note**: As far as I know, this way (explicit) relies always on the **subplots** (with an *s* at the end) method to create graphs. It doesn't matter if it's just only one graph.

#### <a id = 'oosg'>Plotting a single graph with OO-style</a>

```python
fig = plt.figure()

# the way I see plt.sublplots is: plt.figure() + plt.subplot()
fig, ax = plt.subplots(figsize = (6,6),  layout = 'tight')

# now we call on methods
ax.set_xlabel('This is the X label')
ax.set_ylabel("This is the Y label")
ax.set_title("This is the graph title with OO-style")

ax.plot(x, y, label = 'myGraph')
ax.legend()
```

> **Note**: plt.show() could have been used instead.

#### <a id = 'oomg'>Plotting multiple graghs with OO-style</a>

```python
y2 = np.array([2,6,1,4,6,7])

# width and height (in that order)
fig = plt.figure()
fig, ax = plt.subplots(1,2, figsize = (6,4), layout = 'tight') #it's going to have 1 row, 2 cols.

# first graph that we indicate with [0]
ax[0].set_xlabel('This is the X label of the first graph')
ax[0].set_ylabel('This is the Y label of the first graph')
ax[0].set_title('This is the title of the first graph', loc = 'center')
ax[0].grid(color = 'green', ls= ':')

ax[0].plot(x, y)

# second graph that we indicate with [1]
ax[1].set_xlabel('This is the X label of the second graph')
ax[1].set_ylabel('This is the Y label of the second graph')
ax[1].set_title('This is the title of the second graph', loc = 'center')
ax[1].grid(color = 'black', ls= ':')

ax[1].plot(x, y2)

# adde the suptitle to the figure
fig.suptitle('This is the title of the subplot')
```

## <a id = 'rel'>What does Matplotlib say about using one style or another?</a>

> Matplotlib's documentation and examples **use both the OO and the pyplot styles**. In general, **we suggest using the OO style, particularly for complicated plots**, and functions and scripts that are intended to be reused as part of a larger project. However, the **pyplot style can be very convenient for quick interactive work**.
