## <a id = 'comp'>figures vs plots vs graphs</a>
We may get confused about the differences among these words. Let's try to explain what are they exactly:

$$ Figure > Plot > Graph $$

Basically this is saying that a **plot** may contain multiple **graphs**. And a **figure** may contain multiple **plots**.

Before at looking at some examples (in which we are going to see this perfectly clear), let's look at the next python code:

```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.array([1,2,3])
y = np.array([40,50,60])

fig = plt.figure()
fig, ax = plt.subplots(layout = 'tight')
ax.plot(x,y)
```

In this very basic python code we can see that:
- *fig* is the **Figure**
- *ax* is the **Plot**
- *x,y* (data) are the **Graph**

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.array([0,1,2,3])
y = np.array([6,2,6,7])

plt.figure(figsize = (3,3))
plt.plot(x,y)
plt.title('This figure has one plot and one graph')

plt.show()
````

````python
y2 = np.array([0,3,9,2])

plt.figure(figsize = (3,3))
plt.plot(x,y, label ='first graph')
plt.plot(x,y2, label ='second graph')
plt.title('This figure has one plot and two graph')
plt.legend()

plt.show()
````

````python
plt.figure(figsize = (6,3), layout = 'tight')
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('left plot')

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('rigth plot')

plt.suptitle('This figure has two plots with one graph each')

plt.show()
````

````python
y3 = np.array([5,4,4,5])
y4 = np.array([1,0,5,5])

fig = plt.figure()
fig, ax = plt.subplots(1,2, figsize = (6,3), layout = 'tight')

# plot 1
ax[0].set_title('Left plot with two graphs')
ax[0].plot(x,y, label = 'first left graph')
ax[0].plot(x,y2, label = 'second left graph')
ax[0].legend(loc = 'lower right')

#plot 2
ax[1].set_title('Right plot with two graphs')
ax[1].plot(x,y3, label = 'first right graph')
ax[1].plot(x,y4, label = 'second right graph')
ax[1].legend(loc = 'lower right')

fig.suptitle('This figure has two plots with two graphs each')
````

## <a id = 'att'>Attributes</a>
Figure has some simple attributes that we might want to take a look at

### <a id = 'size'>figsize</a>
Modifies the figure's size: **(width, height)**.

>**Note**: For a single graph, a size of (8,12) for big figures and (5,2.7) for small figures are very commonly used.
>**Note**: For multiple graphs, a size of (4,3) is very commonly used.

````python
# big simple graph
plt.figure(figsize = (8,12))
plt.plot(x,y)

plt.show()
````

````python
# big simple graph
plt.figure(figsize = (5,2.7))
plt.plot(x,y)

plt.show()
````

````python
fig = plt.figure()
fig, ax = plt.subplots(1,2,figsize = (4,3))
ax[0].plot(x,y)
ax[1].plot(x,y2)
````

### <a id = 'lay'>Layout</a>
When plotting subplots, it is normal that the titles of the plot get overlapped. To fix this, we can set the layout to:
- *tight*
- *constrained*

````python
# with no layout 
fig = plt.figure()
fig, ax = plt.subplots(1,2)
ax[0].plot(x,y)
ax[0].set_title('This is a very long overlapping title')
ax[1].plot(x,y2)
ax[1].set_title('This is a another long overlapping title')
````

````python
# with tight layout
fig = plt.figure()
fig, ax = plt.subplots(1,2,layout = 'tight')
ax[0].plot(x,y)
ax[0].set_title('This is a very long overlapping title')
ax[1].plot(x,y2)
ax[1].set_title('This is a another long overlapping title')
````

````python
# with constrained layout
fig = plt.figure()
fig, ax = plt.subplots(1,2,layout = 'constrained')
ax[0].plot(x,y)
ax[0].set_title('This is a very long overlapping title')
ax[1].plot(x,y2)
ax[1].set_title('This is a another long overlapping title')
````

### <a id = 'face'>facecolor</a>
Color of the figure's background

````python
plt.figure(facecolor = 'magenta')
plt.plot(x,y)

plt.show()
````

### <a id = 'edge'> edgecolor</a>
Modifies the figure's edgecolor.
>**Note**: the figure's linewidth is = 0 by default so unless you change this value to some other value (2, for example) you won't see any color modifications.

````python
plt.figure(edgecolor = 'fuchsia',linewidth = 2)
plt.plot(x,y)

plt.show()
````

## <a id = 'sub'>Subfigures</a>
The same way you have subplots in a figure, you can have **subfigures inside a figure**. 

- This can be done with the function **subfigures**.

````python
# principal figure
fig = plt.figure(layout='constrained', facecolor='black') #the main figure is going to have black background
fig.suptitle('Figure', color = 'white')

#subfigures declaration. There are going to be 2 subfigures set in 1 row and 2 cols
figL, figR = fig.subfigures(1, 2)

# left subfigure configuration
figL.set_facecolor('thistle')
axL = figL.subplots(2, 1, sharex=True) # multiple plots inside this subfigure(2 rows and 1 col)
axL[1].set_xlabel('x [m]')
figL.suptitle('Left subfigure')

# rigth subfigure configuration
figR.set_facecolor('paleturquoise')
axR = figR.subplots(1, 2, sharey=True) # multiple plots inside this subfigure(1 row and 2 cols)
axR[0].set_title('Axes 1')
figR.suptitle('Right subfigure')
````
